```python
import altair as alt
import pandas as pd
```

```python
percentage_hazardous_asteroids = pd.read_csv('percentage_hazardous_asteroids.csv')

percentage_hazardous_asteroids['safe_asteroids_in_month'] = percentage_hazardous_asteroids['total_asteroids_in_month'] - percentage_hazardous_asteroids['hazardous_count_in_month']

percentage_hazardous_asteroids_long = pd.melt(percentage_hazardous_asteroids, id_vars=['month'], value_vars=['hazardous_count_in_month', 'safe_asteroids_in_month'])

percentage_hazardous_asteroids_long['variable'] = percentage_hazardous_asteroids_long['variable'].replace({
        'hazardous_count_in_month': 'Hazardous',
        'safe_asteroids_in_month': 'Safe'
    })

percentage_hazardous_asteroids_long['stack_order'] = percentage_hazardous_asteroids_long['variable'].map({
    'hazardous_count_in_month': 0,
    'safe_asteroids_in_month': 1
})

alt.Chart(percentage_hazardous_asteroids_long).mark_bar().encode(
    x=alt.X('month', title='Month'),
    y=alt.Y('value', title='Number of Asteroids'),
    color=alt.Color('variable', title='Asteroid Type', scale=alt.Scale(domain=['Hazardous', 'Safe'], range=['red', 'green'])),
    order=alt.Order('stack_order', sort='ascending')
).properties(
    width=600,
    height=400,
    title='Hazardous vs Safe Asteroids per Month'
)
```

```python
top3_closest_asteroids = pd.read_csv('top3_closest_asteroids.csv')

closest_per_month = top3_closest_asteroids[
    top3_closest_asteroids['ranked'] == 1
].sort_values('month')

alt.Chart(closest_per_month).mark_line(point=True).encode(
    x=alt.X('month', title='Month'),
    y=alt.Y('miss_distance_km', title='Distance to Earth (km)')
).properties(
    width=700,
    height=400,
    title='Closest Distance to Earth per Month'
)
```
